```python
import pandas as pd
import numpy as np

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df['season'] = df['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return df

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

```python
def hard_func(df, city_name):
  '''Описание
  1. У вас есть «тяжёлая» функция, которую вы пробуете распараллелить. В ней для одного города:
    1.1. считаются скользящим окном среднее и std. Там же ищутся все аномалии относительно скользящих средних и std. Больше нигде и никогда вы эти данные полученные скользящим окном не используете.
    1.2. Потом вы делаете там же groupby по сезону и считаете mean и std для исходной температуры. Таким образом получаете профиль сезона. Для пункта 2 вам нужны эти mean и std.
    1.3. Ищете тренд (регрессия или что-то получше). Если обучаете регрессию, смотрите на коэффициент и возвращаете положительный или отрицательный тренд.
    1.4. Считаете среднюю, min, max температуры за всё время.
  Итого из функции возвращается:
    Город
    Average, min, max температуры, профиль сезона постороенный со скользящим окном и последующими mean / std и со скользящим окном и без него, а также уклон тренда и список аномальных точек для города.
    '''
  df = df[df['city'] == city_name].copy()
  # скользящее среднее за 30 дней
  df['mean_roll_30'] = df.rolling(window=30, on='timestamp', closed='left')['temperature'].mean()
  # скользящее std за 30 дней
  df['std_roll_30'] = df.rolling(window=30, on='timestamp', closed='left')['temperature'].std()
  # так как в самом первом 30-дневном окне будут наны, то заполняем их первым не nan значением
  df['mean_roll_30'] = df['mean_roll_30'].bfill()
  df['std_roll_30'] = df['std_roll_30'].bfill()
  # верхняя и нижняя граница (+-2std)
  df['upper_bound'] = df['mean_roll_30'] + 2 * df['std_roll_30']
  df['lower_bound'] = df['mean_roll_30'] - 2 * df['std_roll_30']
  # aномалии
  anomalies = df[(df['temperature'] > df['upper_bound']) | (df['temperature'] < df['lower_bound'])]
  # профиль сезона для города
  seasonal_stats = df.groupby('season')['temperature'].agg(['mean', 'std', 'min', 'max']).reset_index()
  # тренд при помощи линейной регрессии
  df['timestamp_ordinal'] = df['timestamp'].map(pd.Timestamp.toordinal)
  X = df['timestamp_ordinal'].values.reshape(-1, 1)
  y = df['temperature']
  model = LinearRegression()
  model.fit(X, y)
  trend = model.coef_[0]
  return {
        "city": city_name,
        "anomalies": anomalies,
        "seasonal_stats": seasonal_stats,
        "trend": "positive" if trend > 0 else "negative" if trend < 0 else "neutral"
        }
```

```python
# визуализация ряда и аномалий
def plot_anomalies(city_data, anomalies, city_name):
    plt.figure(figsize=(12, 6))
    plt.plot(city_data['timestamp'], city_data['temperature'], label='Температура', color='blue', alpha=0.7)
    plt.scatter(anomalies['timestamp'], anomalies['temperature'], color='red', label='Аномалии', zorder=2)
    plt.title(f"Температура и аномалии для города {city}")
    plt.xlabel("Дата", fontsize=10)
    plt.ylabel("Температура")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
cities = list(data.city.unique())
for city in cities:
  result = hard_func(data, city)
  plot_anomalies(data[data['city']==city], result['anomalies'], city)
```

```python
# последовательное выполнение
def analyze_sequentially(data, cities):
    results = []
    for city in cities:
        result = hard_func(data, city)
        results.append(result)
    return results

# параллельное выполнение с ProcessPoolExecutor
def analyze_parallel_pool_executor(data, cities):
    with ProcessPoolExecutor(max_workers=os.cpu_count()) as executor:
        results = []
        futures = [executor.submit(hard_func, data, city) for city in cities]
    for future in futures:
        results.append(future.result())
    return results

# параллельное выполнение с joblib
def analyze_parallel_joblib(data, cities):
    results = Parallel(n_jobs=os.cpu_count())(delayed(hard_func)(data, city) for city in cities)
    return results

# бенчмарк
def benchmark_multiple_runs(func, runs, *args, **kwargs):
    times = []
    for _ in range(runs):
        start_time = time.time()
        func(*args, **kwargs)
        end_time = time.time() - start_time
        times.append(end_time)
    return mean(times)
```

```python
cities = list(data.city.unique())
runs = 5
sequential_time = benchmark_multiple_runs(analyze_sequentially, runs, data, cities)
parallel_time_pool = benchmark_multiple_runs(analyze_parallel_pool_executor, runs, data, cities)
parallel_time_joblib = benchmark_multiple_runs(analyze_parallel_joblib, runs, data, cities)
print(f"Среднее время последовательного выполнения: {sequential_time:.2f} секунд")
print(f"Среднее время параллельного выполнения ProcessPoolExecutor: {parallel_time_pool:.2f} секунд")
print(f"Среднее время параллельного выполнения joblib: {parallel_time_joblib:.2f} секунд")
```

### Сравнение синхронного и асинхронного подходов

```python
import nest_asyncio
nest_asyncio.apply()
```

```python
def sync_benchmark(cities, api_key):
    start_time = time.time()
    for city in cities:
        url = f"http://api.openweathermap.org/data/2.5/weather"
        params = {
            'q': city,
            'appid': api_key,
            'units': 'metric'
        }
        response = requests.get(url, params=params)
        if response.status_code == 200:
            data = response.json()
            print(f"{city}: {data['main']['temp']}")
        else:
            print(f"Ошибка {response.status_code}")
    end_time = time.time() - start_time
    print(f"Синхронный подход занял: {end_time:.2f} секунд")


async def async_benchmark(cities, api_key):
    start_time = time.time()
    async with aiohttp.ClientSession() as session:
        tasks = []
        for city in cities:
            url = f"http://api.openweathermap.org/data/2.5/weather"
            params = {
                'q': city,
                'appid': api_key,
                'units': 'metric'
            }
            tasks.append(session.get(url, params=params))
        responses = await asyncio.gather(*tasks)

        for response in responses:
            if response.status == 200:
                data = await response.json()
                print(f"{data['name']}: {data['main']['temp']}")
            else:
                print(f"Ошибка {response.status}: {await response.text()}")
    end_time = time.time() - start_time
    print(f"Асинхронный подход занял: {end_time:.2f} секунд")
```

```python
# сравнение синхронного и асинхронного подходов на 20ти запросах
cities = ["Berlin", "Cairo", "Dubai", "Beijing", "Moscow"] * 4
print("20 синхронных запросов:")
sync_benchmark(cities, api_key)
print("-"*50)
print("20 асинхронных запросов:")
asyncio.run(async_benchmark(cities, api_key))
```

Как и ожидалось, асинхронный подход работает намного быстрее.

### Проверка температур на аномальность

```python
async def get_current_temperature_async(city, api_key):
    url = f"http://api.openweathermap.org/data/2.5/weather"
    params = {
        'q': city,
        'appid': api_key,
        'units': 'metric'
    }
    async with aiohttp.ClientSession() as session:
        async with session.get(url, params=params) as response:
            if response.status == 200:
                data = await response.json()
                return data['main']['temp']
            else:
                print(f"Ошибка: {response.status}")
                return None

def check_temperature(city, temp, season, seasonal_stats):
    stats = seasonal_stats[seasonal_stats['season'] == season]
    mean = stats['mean'].iloc[0]
    std = stats['std'].iloc[0]
    lower_bound = mean - 2 * std
    upper_bound = mean + 2 * std
    if lower_bound <= temp <= upper_bound:
        return f"Температура в {city} {temp} в цельсиях нормальная для сезона {season}."
    else:
        return f"Температура в {city} {temp} в цельсиях аномальная для сезона {season}!"

async def main():
    cities = ["Berlin", "Cairo", "Dubai", "Beijing", "Moscow"]
    # зима, холода, одинокие ветра(((
    current_season = "winter"
    # статистика для каждого города через реализованную в самом начале hard_func
    seasonal_stats_dict = {
        city: hard_func(data, city)['seasonal_stats']
        for city in cities}
    tasks = []
    for city in cities:
        temp_task = get_current_temperature_async(city, api_key)
        tasks.append(temp_task)

    current_temps = await asyncio.gather(*tasks)

    for city, temp in zip(cities, current_temps):
        if temp is not None:
            seasonal_stats = seasonal_stats_dict[city]
            result = check_temperature(city, temp, current_season, seasonal_stats)
            print(result)

asyncio.run(main())
```
